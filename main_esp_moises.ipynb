```python
#tu codigo aqu铆
# El elemento m谩s qu猫?
# Entiendo que recibe una lista en lugar de un par de enteros
def greatest(values: list[int]) -> int|None:
    """ Retorna el valor mayor de la lista. En caso de lista vacia, retorna None"""
    if not values:
        return None
    major = values[0]
    for i in values[1:]:
        if i > major:
            major = i
    return major
```

## 6. Escribe una funci贸n que devuelva el factorial de un n煤mero.

```python
def factorial(n):
    """ Retorna el factorial de n """
    if n <= 0:
        raise ValueError("V谩lido para valores >= 1")
    if n <= 2:
        return n
    return n * factorial(n - 1)
#tu codigo aqu铆
```

`NOTE: No podemos usar set. `

```python
def unique(lst_un: list):
    """ Retorna la lista con valores 煤nicos. Asume no None """
    result = []
    for i in lst_un:
        if i not in result:
            result.append(i)
    return result
    
```

## 8. Escribe una funci贸n que devuelva la moda de una lista, es decir: el elemento que aparece m谩s veces.

`NOTE: No se debe usar count... `

```python

def mode_counter(arr: list):
    """ retorna la moda (una llista amb els valors m茅s freq眉ents).
    Asume no None i retorna buida si arr 茅s buit """
    def compute_freq(arr: list) -> dict:
    """ Calcula la freq眉猫ncia dels valors en arr. """
    freq = {}
    for i in arr:
        if i not in freq:
            freq[i] = 0
        freq[i] += 1
    return freq
        
    def find_greater(freq: dict):
        """ retorna les claus de la freq amb valor m茅s gran.
        Retorna la llista buida si no n'hi ha elements """
        winner_keys = []
        winner_val = 0
        for key, val in freq.items():
            if val > winner_val:
                winner_keys = [key]
                winner_val = val
            elif val == winner_val: # sharing mode
                winner_keys.append(key)
        return winner_keys

    if not arr:
        return None
    return find_greater(compute_freq(arr))
    
    
```

```python
# Para testear tu funci贸n 
#test_mode(mode_counter)
assert mode_counter([1, 2, 3, 4]) == [1, 2, 3, 4]
assert mode_counter([1, 2, 1, 4]) == [1]
assert mode_counter([1, 2, 1, 2, 3, 2, 4]) == [2]
assert mode_counter([1, 2, 1, 2, 3, 2, 4, 1]) == [1, 2]
```

## 10. Escribe una funci贸n para verificar si una cadena es un pangrama, es decir: si contiene todas las letras del alfabeto al menos una vez. Ten en cuenta que las cadenas pueden contener caracteres que no son letras.

```python
def pangram(string: str) -> bool:
    """ Retorna True si string 茅s un pangrama """
    lletres = unique([ch for ch in string.lower() if ord('a') <= ord(ch) <= ord('z')])
    return len(lletres) == ord('z') - ord('a') + 1
    
```

```python
# Para testear tu funci贸n
#test_pangram(pangram)
totes = "".join(chr(i) for i in range(ord('a'), ord('z')+1))
assert pangram(totes)
totes_amb_nums = "".join(chr(i) for i in range(ord('0'), ord('z')+1))
assert pangram(totes_amb_nums)
no_totes = "".join(chr(i) for i in range(ord('a'), ord('z')+1) if i != ord('j'))
assert not pangram(no_totes)
```

## 11. Escribe una funci贸n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab茅ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! お`

```python
def sort_alpha(string):
    """ Retorna les paraules separades per coma a string, ordenades """
    def meu_split(string: str) -> list[str]:
        """ 茅s el meu split """
        paraules = []
        paraula = []
        for ch in string:
            if ch == ',':
                if paraula:
                    paraules.append("".join(paraula))
                    paraula = []
            else:
                paraula.append(ch)
        if paraula:    # la darrera
            paraules.append("".join(paraula))
        return paraules

    def meu_join(paraules: list[str]) -> str:
        """ 茅s el meu join """
        if not paraules:
            return ""
            
        resultat = paraules[0]
        for paraula in paraules[1:]:
            resultat += f",{paraula}"
        return resultat

    return meu_join(sorted(meu_split(string)))
```

```python
# Para testear tu funci贸n 
#test_alpha(sort_alpha)
assert sort_alpha("undostres") == "undostres"
assert sort_alpha("un,dos,tres") == "dos,tres,un"
assert sort_alpha("un,dos,quatre,dos,tres") == "dos,dos,quatre,tres,un"
```

## 12. Escribe una funci贸n para verificar si una contrase帽a dada es fuerte (al menos 8 caracteres, al menos una min煤scula, al menos una may煤scula, al menos un n煤mero y al menos un car谩cter especial). Deber铆a devolver True si es fuerte y False si no lo es.

```python
def check_pass(password):
    """ Retorna True si el password 茅s fort """
    def has_lower(password: str) -> bool:
        """ returns True if has at least one lowercase char """
        for ch in password:
            if ch.islower():
                return True
        return False
    def has_upper(password: str) -> bool:
        """ returns True if has at least one uppercase char """
        for ch in password:
            if ch.isupper():
                return True
        return False
    def has_digit(password: str) -> bool:
        """ returns True if has at least one digit """
        for ch in password:
            if ord('0') <= ord(ch) <= ord('9'):   # isdigit
                return True
        return False
    def has_special_char(password: str) -> bool:
        """ returns True if has at least one special char
        Special char are non letters a-zA-Z nor digits """
        for ch in password:
            if ord(ch) < ord('0') or ord('9') > ord(ch.upper()) > ord('A') or ord(ch.lower()) > ord('z'):
                return True
        return False
    if len(password) < 8:
        return False
    if not has_lower(password):
        return False
    if not has_upper(password):
        return False
    if not has_digit(password):
        return False
    if not has_special_char(password):
        return False
    return True
    
```
